3. Perform simple arithmetic:


```python
num1 = 45
num2 = 54
sum = num1 + num2
minus = num1 - num2
multiple = num1 * num2
divide = num1 / num2
print("Sum = ",sum)
print("Minus = ",minus)
print("Multiple = ",multiple)
print("Divide = ",divide)
```
